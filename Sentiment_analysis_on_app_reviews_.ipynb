```python
Df = pd.read_csv('reviews.csv')
Df.head()
```

```python
Df = Df[['content', 'sentiment']]
Df.head()
```

```python
import spacy
nlp=spacy.blank("en")
```

```python
# removing stopwords
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
stop = stopwords.words('english')
Df['content_without_stopwords'] = Df['content'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop)]))
Df.head()
```

```python
# remove hyperlinks
Df['content_without_stopwords'] = Df['content_without_stopwords'].replace(to_replace=r'^https?:\/\/.*[\r\n]*',value='',regex=True)
Df.head()
```

```python
# remove numbers and digits
Df["content_without_stopwords"] = Df['content_without_stopwords'].str.replace('[^\w\s]','')
Df.head()
```

```python
# remove punctuations
Df.content_without_stopwords = Df.content_without_stopwords.str.replace('\d+', '')
Df.head()
```

```python
# removing special characters
Df.content_without_stopwords = Df.content_without_stopwords.replace('\*', '', regex=True)
Df.head()
```

```python
# remove emoji
! pip install emoji
import emoji
Df['content_without_stopwords'] = Df['content_without_stopwords'].apply(lambda s: emoji.replace_emoji(s, ''))
Df.head()
```

```python
from nltk.tokenize import word_tokenize
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report
```

```python
content_without_stopwords=Df['content_without_stopwords'].tolist()
sentiment=Df['sentiment'].tolist()
```

```python
nltk.download('punkt')
content_without_stopwords = [word_tokenize(text.lower()) for text in content_without_stopwords]
```

```python
vectorizer = TfidfVectorizer(max_features=5000)
X = vectorizer.fit_transform([' '.join(tokens) for tokens in content_without_stopwords])
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, sentiment, test_size=0.2, random_state=42)
```

```python
classifier = MultinomialNB()
classifier.fit(X_train, y_train)
y_pred = classifier.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
print("Classification Report:")
print(classification_report(y_test, y_pred))
```
